```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

uk_gene_df=pd.read_csv("sim_av_gene.csv")
uk_tumour_df=pd.read_csv("sim_av_tumour.csv")
uk_patient_df=pd.read_csv("sim_av_patient.csv")
```

```python
gene_counts = uk_gene_df.groupby('GENE_DESC')['PATIENTID'].nunique().reset_index()

# Sort the gene types by patient count in descending order
top_gene_types = uk_gene_counts.sort_values(by='GENE_DESC', ascending=False).head(5)

# Display the top 5 gene types
print("Top 5 Gene Types:")
print(top_gene_types)
```

```python
merged_data = uk_tumour_df.merge(uk_patient_df, on='PATIENTID', how='inner')
lung_cancer_data = merged_data[merged_data['SITE_ICD10_O2_3CHAR'] == 'C34']
#death_causes = lung_cancer_data[['PATIENTID', 'SITE_ICD10_O2_3CHAR','DEATHCAUSECODE_1A', 'DEATHCAUSECODE_1B', 'DEATHCAUSECODE_1C', 'DEATHCAUSECODE_2', 'DEATHCAUSECODE_UNDERLYING']]
#death_causes = lung_cancer_data[['PATIENTID', 'SITE_ICD10_O2_3CHAR','DEATHCAUSECODE_UNDERLYING']]
death_causes = lung_cancer_data[['PATIENTID','DEATHCAUSECODE_UNDERLYING']]
cause_counts = death_causes.drop(columns=['PATIENTID']).stack().value_counts()

top_10_causes = cause_counts.head(10)

# Create a bar plot
plt.figure(figsize=(12, 6))
top_10_causes.plot(kind='bar')
plt.title('Top 10 Lung Cancer Death Causes')
plt.xlabel('Death Cause')
plt.ylabel('Patient Count')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```
