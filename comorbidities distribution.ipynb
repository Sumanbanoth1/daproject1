```python
import pandas as pd
dataset = pd.read_csv('sim_av_patient.csv')
dataset1 = pd.read_csv('sim_av_tumour.csv')
```

```python
import pandas as pd



# Merge the two DataFrames based on the 'PATIENTID' column
merged_dataset = pd.merge(dataset, dataset1, on='PATIENTID', how='inner')
```

```python
sorted_dataset = merged_dataset.sort_values(by='PATIENTID', ascending=True)
```

```python
sorted_dataset.head(10)
```

```python
required_columns = sorted_dataset[['PATIENTID', 'COMORBIDITIES_27_03', 'VITALSTATUS']]
```

```python
required_columns.head(10)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# DataFrame named 'required_columns' with the relevant data

# Step 1: Filter the data for deceased patients (VITALSTATUS 'D')
deceased_patients = required_columns[required_columns['VITALSTATUS'] == 'D']

# Step 2: Create a new column 'Comorbidity_Group' based on the first two characters of 'COMORBIDITIES_27_03'
deceased_patients['Comorbidity_Group'] = deceased_patients['COMORBIDITIES_27_03'].str[:2]

# Step 3: Count the comorbidity groups
comorbidity_counts = deceased_patients['Comorbidity_Group'].value_counts().sort_index()

# Step 4: Create a bar chart to visualize the distribution with a legend
plt.figure(figsize=(12, 6))
comorbidity_counts.plot(kind='bar', label='Comorbidity Groups')
plt.title('Distribution of Comorbidity Groups in Deceased Patients')
plt.xlabel('Comorbidity Groups')
plt.ylabel('Count')
plt.xticks(rotation=0)  

# Add a legend
plt.legend()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a dictionary to map group codes to disease names
comorbidity_mapping = {
    '01': 'Acute Myocardial Infarction',
    '02': 'Congestive Heart Failure',
    '03': 'Peripheral Vascular Disease',
    '04': 'Cerebral Vascular Accident',
    '05': 'Dementia',
    '06': 'Pulmonary Disease',
    '07': 'Connective Tissue Disorder',
    '08': 'Peptic Ulcer',
    '09': 'Diabetes',
    '10': 'Diabetes Complications',
    '11': 'Paraplegia',
    '12': 'Renal Disease',
    '13': 'Cancer',
    '14': 'Metastatic Cancer',
    '15': 'Liver Disease',
    '16': 'HIV',
    '17': 'Severe Liver Disease'
}

# DataFrame named 'required_columns' with the relevant data

# Step 1: Filter the data for deceased patients (VITALSTATUS 'D')
deceased_patients = required_columns[required_columns['VITALSTATUS'] == 'D']

# Step 2: Create a new column 'Comorbidity_Group' based on the first two characters of 'COMORBIDITIES_27_03'
deceased_patients['Comorbidity_Group'] = deceased_patients['COMORBIDITIES_27_03'].str[:2]

# Step 3: Count the comorbidity groups
comorbidity_counts = deceased_patients['Comorbidity_Group'].value_counts().sort_index()

# Step 4: Create a bar chart to visualize the distribution with a legend and values
plt.figure(figsize=(12, 6))
bars = comorbidity_counts.plot(kind='bar', label='Comorbidity Groups')
plt.title('Distribution of Comorbidity Groups in Deceased Patients')
plt.xlabel('Comorbidity Groups')
plt.ylabel('Count')
plt.xticks(rotation=0)  

# Add a legend based on the provided mapping
legend_labels = [f'{group}: {disease}' for group, disease in comorbidity_mapping.items()]
plt.legend(handles=[plt.Line2D([0], [0], color='b', lw=2, label=label) for label in legend_labels])

# Add values to the bars
for bar in bars.patches:
    height = bar.get_height()
    plt.annotate(f'{height}', (bar.get_x() + bar.get_width() / 2, height), ha='center', va='bottom')

plt.show()
```
